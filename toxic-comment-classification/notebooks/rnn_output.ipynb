```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../data/cross_validation/291367/rnn.csv')
df.sort_values('auc', ascending=False)
```

```python
sns.swarmplot(x='bidirectional', y='auc', data=df)
plt.ylim([0.986, 0.990])
```

```python
sns.boxplot(x='bidirectional', y='auc', data=df)
```

```python
sns.swarmplot(x='lstm_last', y='auc', hue='bidirectional', data=df)
plt.ylim([0.986, 0.990])
```

```python
sns.boxplot(x='lstm_last', y='auc', data=df)
```

```python
sns.swarmplot(x='lstm_max', y='auc', hue='bidirectional', data=df)
plt.ylim([0.986, 0.990])
```

```python
sns.boxplot(x='lstm_max', y='auc', data=df)
```

```python
sns.swarmplot(x='lstm_mean', y='auc', hue='bidirectional', data=df)
plt.ylim([0.986, 0.990])
```

```python
sns.boxplot(x='lstm_mean', y='auc', data=df)
```
