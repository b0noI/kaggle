```python
import random
from datetime import date

import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline
```

```python
data = pd.read_csv('../input/train_1.csv')
data = pd.melt(data, id_vars='Page', var_name='Date', value_name='Views')
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.head()
```

```python
pages = data['Page'].unique()
'{:,}'.format(len(pages))
```

```python
for k in range(20):
    page = random.choice(pages)
    page_data = data[data['Page'] == page] 
    plt.figure(figsize=(16, 6))
    plt.plot(page_data['Date'], page_data['Views'], color='blue')
    for y in (2015, 2016):
        plt.axvspan(*mdates.date2num([date(y, 9, 13), date(y, 11, 13)]), color='red', alpha=0.1)
    plt.title(page)
    plt.grid()
```
